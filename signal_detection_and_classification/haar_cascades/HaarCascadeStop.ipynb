 ## Import Libraries

```python
from PIL import Image
import numpy as np
import threading
import time
import cv2
import io
import matplotlib.pyplot as plt
import math
from math import sqrt
from skimage.feature import blob_dog, blob_log, blob_doh
from scipy import misc
from scipy import ndimage
```

 ## Haar classifier

```python
classifier = cv2.CascadeClassifier('C:/Users/vhar/Desktop/TFMFinal/SinalCascade/Stop/cascade.xml')
```

 ## Test Image

```python
imagen = cv2.imread('C:/Users/vhar/Desktop/TFMFinal/SinalCascade/Stop/img/10cm.jpg')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.show()
```

 ## Detect Object

```python
detectado = False
imagen = cv2.GaussianBlur(imagen,(5,5),0)
cascade_obj = classifier.detectMultiScale(imagen,scaleFactor=1.1,minNeighbors=5,minSize=(20, 20),flags=cv2.CASCADE_SCALE_IMAGE)
print('Numero de Objetos: ',len(cascade_obj))
if (len(cascade_obj)):
    cas = np.array([cascade_obj[0]])
    detectado = True
    for (x_pos, y_pos, width, height) in cas:
        print('X: ',x_pos)
        print('Y: ',y_pos)
        print('Ancho: ',width)
        print('Alto: ',height)
        cv2.rectangle(imagen, (x_pos+(-10), y_pos+(-10)), (x_pos+width-(-10), y_pos+height-(-10)), (255, 255, 255), 2)
        recorte = imagen[y_pos+(-10):y_pos+height-(-10),x_pos+(-10):x_pos+width-(-10)]
        cv2.imwrite('C:/Users/vhar/Desktop/TFMFinal/SinalCascade/Stop/img/stop.jpg',recorte)
        cv2.putText(imagen, 'STOP', (x_pos, y_pos-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
else:
    detectado = False
print('Detectado: ',detectado)
```

 ## Result

```python
fig = plt.figure()

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(recorte, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
